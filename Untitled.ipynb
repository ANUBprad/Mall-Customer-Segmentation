```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
# Load data
data = pd.read_csv('Mall_Customers.csv')
print(f"Data shape: {data.shape}")
data.head()
```

```python
# Exploratory Data Analysis
print("Basic statistics:")
print(data.describe())

print("\nGender distribution:")
print(data['Gender'].value_counts())
```

```python
# Visualize distributions
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution')

plt.subplot(1, 3, 2)
sns.histplot(data['Annual Income (k$)'], bins=20, kde=True)
plt.title('Income Distribution')

plt.subplot(1, 3, 3)
sns.histplot(data['Spending Score (1-100)'], bins=20, kde=True)
plt.title('Spending Score Distribution')

plt.tight_layout()
plt.show()
```

```python
# Select features and scale
features = ['Annual Income (k$)', 'Spending Score (1-100)']
X = data[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Determine optimal clusters using Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Perform clustering with optimal clusters (5)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_scaled)
data['Cluster'] = clusters
```

```python
# Visualize clusters
plt.figure(figsize=(12, 8))
for i in range(n_clusters):
    plt.scatter(
        X[clusters == i]['Annual Income (k$)'], 
        X[clusters == i]['Spending Score (1-100)'], 
        label=f'Cluster {i}', 
        s=50
    )

plt.scatter(
    kmeans.cluster_centers_[:, 0], 
    kmeans.cluster_centers_[:, 1], 
    s=200, c='black', marker='X', label='Centroids'
)
plt.title('Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
# Analyze clusters
cluster_summary = data.groupby('Cluster').agg({
    'Age': 'mean',
    'Annual Income (k$)': 'mean',
    'Spending Score (1-100)': 'mean',
    'Gender': lambda x: x.mode()[0],
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'Count'})

print("Cluster characteristics:")
cluster_summary
```

```python
# 3D visualization (optional)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

features_3d = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
X_3d = data[features_3d]

for i in range(n_clusters):
    ax.scatter(
        X_3d[clusters == i]['Age'],
        X_3d[clusters == i]['Annual Income (k$)'],
        X_3d[clusters == i]['Spending Score (1-100)'],
        label=f'Cluster {i}',
        s=50
    )

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.title('3D Cluster Visualization')
plt.legend()
plt.show()
```

```python
# Save model for deployment
import joblib

joblib.dump(kmeans, 'kmeans_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("Model and scaler saved for deployment")
```
